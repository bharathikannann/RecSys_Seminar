## This notebook is used to visualize the performance of feature selection methods 

```python
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
path = 'results-from-baseline/'
```

```python
df = pd.read_csv(path + "chi2.csv", header = None, names=['Chi-2'])
df["F-classif"] = pd.read_csv(path + "f-classif.csv", header = None)
df["F-regression"] = pd.read_csv(path + "f-regression.csv", header = None)
df["Mutual Information Regression"] = pd.read_csv(path + "mutual_importance_regression.csv", header = None)
df["Variance Threshold"] = pd.read_csv(path + "Variance-threshold.csv", header = None)
df['Permutation Importance'] = pd.read_csv(path + "permutation_importance.csv", header = None)
df["Recursive Feature Elimination"] = pd.read_csv(path + "recursive_feature_elimination.csv", header = None)
df['Random Forest'] = pd.read_csv(path + "random_forest.csv", header = None)
df["Extra Tree Classifier"] = pd.read_csv(path + "extra_tree_classifier.csv", header = None)
df["Select from Model (LR)"] = pd.read_csv(path + "select_from_model.csv", header = None)
```

## Plot the graph

```python
colors = ['olive','m','y','brown','b','c','g','orange','r','sienna']
plt.style.use('default')
plt.figure(figsize=(12,8))
plt.ylim(0.885,0.899)
plt.xlim(-50, 550)
plt.plot(522, 0.8928217186218789, marker="o", markersize=10, markeredgecolor="r", markerfacecolor="r", label="Baseline Model Performance")
for c, i in enumerate(range(0,len(df.columns))):
    if df.iloc[:,i].name == "Select from Model (LR)":
        plt.plot([0,50,100,150,196,196,196,196,196,196,196], df.iloc[:,i].values[1:-1], color = colors[i], alpha=0.7, label = df.iloc[:,i].name, linestyle = '--', marker = '*')
    else:
        plt.plot([0,50,100,150,200,250,300,350,400,450,500, 522], df.iloc[:,i].values[1:], color = colors[i], label = df.iloc[:,i].name, linestyle = '--', marker = '*')
plt.plot(0, 0.897811, marker="o", markersize=7, markeredgecolor="b", markerfacecolor="b", label="Model with no add features")
plt.xlabel("No of Additional Parameters")
plt.ylabel("Validation RMSE")
plt.legend(bbox_to_anchor=(1.31,1))
plt.show()
```
